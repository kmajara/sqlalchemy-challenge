```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, text, func, inspect, and_
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model by: 

#creating a base class for refletion
Base = automap_base()

# reflect the tables
Base.prepare(autoload_with=engine)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
inspector = inspect(engine)
```

```python
# Inspect columns in the Measurement table
columns = inspector.get_columns('Measurement')

for c in columns:
    print(c['name'], c['type'])
```

```python
# Find the most recent date in the data set

session.query(Measurement.date).order_by(Measurement.date.desc()).first()
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
latest_date = (dt.date(2017, 8, 23))

# Calculate the date one year from the last date in data set [assuming 365 days]
one_year_ago = latest_date - dt.timedelta(days=365)

# Perform a query to retrieve the data and precipitation scores
query_result = session.query(Measurement.date, Measurement.prcp).\
    filter(and_(Measurement.date >= one_year_ago, Measurement.date <= latest_date)).\
    order_by(Measurement.date.asc()).all()

# Save the query results as a Pandas DataFrame. Explicitly set the column names
precip_results_df = pd.DataFrame(query_result)
new_column_names = {'date': 'Date', 'prcp': 'Precipitation'}
precip_results_df.rename(columns=new_column_names, inplace=True)

# Sort the dataframe by date
precip_results_df = precip_results_df.sort_values(by='Date')
precip_results_df.head(15)

# Use Pandas Plotting with Matplotlib to plot the data
#precip_results_df.plot(x='Date', y='Precipitation', marker='o', linestyle='-')

plt.figure(figsize=(12, 6))
plt.plot(precip_results_df['Date'], precip_results_df['Precipitation'], linestyle='-')
plt.xlabel('Date')
plt.ylabel('Inches')
plt.title('Precipitation Values by Day Over One Year')
plt.xticks(rotation = 90)
plt.tight_layout()
plt.grid(True)

# Set x-axis tick labels
num_dates_to_display = 5
date_indices = range(0, len(precip_results_df), len(precip_results_df) // num_dates_to_display)
plt.xticks(precip_results_df['Date'].iloc[date_indices], rotation=90)
plt.legend(["Precipitation"])
# plt.tight_layout()
plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data

summary_stats = precip_results_df.describe()
summary_stats
```

# Exploratory Station Analysis

```python
#latest date
latest_date = dt.date(2017, 8, 23)

# One year before the latest date
one_year_ago = latest_date - dt.timedelta(days=365)

# Find the most active station
most_active_station = session.query(Measurement.station).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.id).desc()).first()[0]

# Find tobs for the most active station within the last year
active_tobs = session.query(Measurement.date, Measurement.tobs).\
    filter(Measurement.station == most_active_station).\
    filter(and_(Measurement.date >= one_year_ago, Measurement.date <= latest_date)).\
    order_by(Measurement.date.asc()).all()

session.close()
```

```python
#inspect columns included in the Station table
columns = inspector.get_columns('Station')

for c in columns:
    print(c['name'], c['type'])
```

```python
# Design a query to calculate the total number of stations in the dataset

# total_stations = session.query(Station).distinct().count()
station_count = [(session.query(func.count(Station.station)).scalar(),)]

# print(total_stations)
station_count
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# List the stations and their counts in descending order.
active_stations = session.query(Measurement.station,func.count(Measurement.id)) \
                         .group_by(Measurement.station) \
                         .order_by(func.count(Measurement.id).desc()) \
                         .all()
#print the active stations
active_stations
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.

aggregate_list = [func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)]
most_active_data = session.query(*aggregate_list).group_by(Measurement.station).order_by(func.count(Measurement.id).desc()).first()

print(most_active_data)
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
# per review of the CSV file, the most recent date associated with the most active station is 8/17/2017

# create date parameters
query_date = dt.datetime(2017, 8, 18)
one_year_ago = query_date - dt.timedelta(days = 365)

#create the list of temperatures for the most active station based on the date
temp_query_result = session.query(Measurement.tobs).filter(Measurement.station == active_stations[0][0])\
                                                    .filter(Measurement.date >= one_year_ago).all()
#create a dataframe from the data 
temp_query_result_df = pd.DataFrame(temp_query_result)

#Create a histogram from the data
temp_query_result_df.hist(bins=12, legend = True)
plt.xlabel('Frequency')
plt.ylabel('Temperature')
```
